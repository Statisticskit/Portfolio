```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
pd.set_option('max_colwidth', 100)
```

```python
nutrition_data = pd.read_csv(r"C:\Users\kitty\Projects\McDonalds Nutrition\McDonaldsMenuNutrition.csv")

# display dataframe
nutrition_data
```

```python
# view column names in dataframe to decide what we're interested in
nutrition_data.columns
```

```python
# Create subset with macronutrients
nutrition_data_subset = nutrition_data[['Item', 'Calories', 'Total Fat\n(g)', 'Carbs\n(g)','Protein\n(g)']]

#view dataframe subset
nutrition_data_subset
```

```python
# check for any null values
nutrition_data_subset.isna().sum()
```

```python
# look for rows where col value is NA
nutrition_data_subset[nutrition_data_subset['Calories'].isna()]
```

Here I can see the rows with the missing values are all on the same items

```python
# numerical summary of data subset
nutrition_data_subset.describe()
```

```python
# look at the type of vectors
nutrition_data_subset.dtypes
```

```python
# create histograms of numerical data
nutrition_data_subset.hist(bins=60, figsize=(20,10))
```

```python
# create df where the calories are above 700
nutrition_data_subset_high_cal = nutrition_data_subset[nutrition_data_subset['Calories']>700]

#create plot of items with high calories 
nutrition_data_subset_high_cal[['Item', 'Calories']].plot(kind='bar', x='Item', y='Calories', figsize=(20,10))
```

Here we can see the items with the highest calories, we can do this for any of the other macro nutrients of we want to

```python
#create scatter plot of items with high calories and their fat contents
nutrition_data_subset_high_cal[['Calories', 'Total Fat\n(g)']].plot(kind='scatter', x='Calories', y='Total Fat\n(g)', figsize=(20,10))
```
